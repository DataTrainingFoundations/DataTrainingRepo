```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

#Load the Tips dataset
tips = sns.load_dataset("tips")

sns.histplot(data=tips, x= "total_bill")



```

```python
tips.head()
```

```python
#changing the bin range could create a very different looking plot
bins = np.arange(0,60,10)
sns.histplot(data=tips,x="total_bill",stat="density",bins=bins,color="orange",edgecolor="blue",kde=True)
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips)
```

```python
#customize scatter plot


sns.scatterplot(x="total_bill",y="tip",hue="sex",size="size",sizes=(50,200), data=tips)
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title("Relationship between Total Bill and Tip")
```

```python
#create a density plot of the "total_bill" column from "tips"
# we will use "hue" parameter to fill the area under the cure
# we adjust the "alpha" and "linewidth" parameters to make the plot more visually appealing
sns.kdeplot(data=tips,x="total_bill",hue="time",fill=True, alpha=0.6,linewidth=1.5)
```

```python
tips.describe()
```

```python
#boxplots are a visual of the describe() method essentially
#plt.figure(figsize=(12,6))
#fig, ax = plt.subplots(figsize=(12,6))
#ax.set_xlim(-0.5,len(tips['day'].unique())-0.5)
sns.boxplot(x="day", y="total_bill",data=tips, hue="day", width=0.8, dodge=False)
plt.title("Total Bill Each Day")

```

```python
sns.violinplot(x="day",y="total_bill",data=tips,hue="day", width=0.8, dodge=False)
plt.title("Total Bill Each Day")
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x="day",y="total_bill",hue="time", data=tips, palette="Set3",linewidth=1.5, fliersize=10)
```

```python
#create a pair plot
sns.pairplot(data=tips, hue="size",diag_kind="kde",palette="husl")
```

```python
jp=sns.jointplot(x="tip",y="total_bill", data=tips,hue="day")
#y>1 moves up y<1 moves down
jp.figure.suptitle("Joint Plot of Total Bill and Tips",fontsize=16, y=1.03)
#plt.title("Joint Plot of Total Bill and Tips")
```

```python
#Facet grids if we have categorical variables nad two other variables of interest
#create a FacetGrid for size vs total_bill
fg = sns.FacetGrid(tips,col="size")

#plot histogram for total_bill in each size category
fg.map(sns.histplot,"total_bill")
```

```python
#Facet grids if we have categorical variables nad two other variables of interest
#create a FacetGrid for size vs total_bill
fg = sns.FacetGrid(tips,col="time",)

#plot histogram for total_bill in each size category
fg.map(sns.histplot,"total_bill")

titles = ["Lunch" if x==False else "Dinner" for x in fg.col_names]
for ax, title in zip(fg.axes.flat, titles):
    ax.set_title(title)

plt.text(40,15,"More money \n is spent \n at dinner",fontsize=12,color="green")

fg.figure.suptitle("Total Bill at different Meals",fontsize=16, y=1.04)

```
