```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

#Load the Tips dataset
tips = sns.load_dataset("tips")

sns.histplot(data=tips, x= "total_bill")



```

```python
tips.head()
```

```python
#changing the bin range could create a very different looking plot
bins = np.arange(0,60,10)
sns.histplot(data=tips,x="total_bill",stat="density",bins=bins,color="orange",edgecolor="blue",kde=True)
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips)
```

```python
#to see all the variables
tips.columns
```

```python
#customize scatter plot


sns.scatterplot(x="total_bill",y="tip",hue="sex",size="size",sizes=(50,200), data=tips)
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title("Relationship between Total Bill and Tip")
```

```python
tips["size"].describe()
```

```python
tips["size"].unique()
```

```python
#create a density plot of the "total_bill" column from "tips"
# we will use "hue" parameter to fill the area under the cure
# we adjust the "alpha" and "linewidth" parameters to make the plot more visually appealing
sns.kdeplot(data=tips,x="total_bill",hue="time",fill=True, alpha=0.6,linewidth=1.5)
```

```python
tips.describe()
```

```python
#boxplots are a visual of the describe() method essentially
sns.boxplot(x="day", y="total_bill",data=tips, hue="day", width=0.8)
plt.title("Total Bill Each Day")
```
